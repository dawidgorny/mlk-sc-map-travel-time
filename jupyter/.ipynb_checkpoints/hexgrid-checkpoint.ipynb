## Load generated hexgrid

```python
with open("../dist/hexgrid_source.geojson", "r") as read_file:
    data = json.load(read_file)
```

```python
print(json.dumps(data["features"][:2], indent=2, sort_keys=True))
```

```python
cellId = 1
for f in data["features"]:
    f["properties"]["cell-id"] = cellId
    f["properties"]["address_count"] = 1
    f["properties"]["district_full_text"] = ""
    cellId += 1
```

```python
print(json.dumps(data["features"][:2], indent=2, sort_keys=True))
```

## Output

```python
with open("../dist/test.geojson", "w") as write_file:
    json.dump(data, write_file)
```
